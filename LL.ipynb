```python
# Libraries
import pandas as pd
import numpy as np
import statsmodels.formula.api as smf

# Data manipulating
LL = pd.read_stata('LL_train.dta')
pd.set_option('display.max_columns', 50)
pd.set_option('display.max_rows', 10)

## create income per million variable
## # data['newvar'] = data.oldvar / 1000000
LL['income_mill'] = LL.income_month / 1000000
```

```python
# Y = income_month
# OLS Regression
model1 = smf.ols('income_month ~ age + agesqr + female', data=LL).fit()

# OLS Regression with quadratic variable (assuming agesqr is not exist)
model2 = smf.ols('income_month ~ age + I(age**2) + female', data=LL).fit()


# Y = income_mill
# OLS Regression
model3 = smf.ols('income_mill ~ age + agesqr + female', data=LL).fit()

# OLS Regression with quadratic variable (assuming agesqr is not exist)
model4 = smf.ols('income_mill ~ age + I(age**2) + female', data=LL).fit()
```

```python
# Result [income_month]
# please compare it yourself
print(model1.summary())
print(model2.summary())
```

```python
# Result [income_mill]
# please compare it yourself
print(model3.summary())
print(model4.summary())
```
